```python
from Amrith_lib import *
import matplotlib.pyplot as plt
import numpy as np
```

```python
def r(x,y):
    return (x**2 + y**2)

def find_pi(num):
    temp=lcg_rng(x0=0.1,n=2*num,range=[0,1])
    x=temp[0:num-1]
    y=temp[num:2*num -1]
    count=0
    for i in range(len(x)):
        if r(x[i], y[i])<=1:
            count+=1
    
    return 4*(count/num)


n_list=np.arange(20,2000,10)
pi_list=[find_pi(i) for i in n_list]
plt.plot(n_list,pi_list)
plt.xlabel("n")
plt.ylim(0,4)
plt.ylabel(r"value of $\pi$")
plt.axhline(3.14,c='r')
plt.grid(True)
plt.show()

pi=np.average(pi_list[-len(pi_list)//10:-1])
print(f"the value of pi is {pi}")
```
