# Question-1

```python
def f(x):
    return np.log10(x/2)-np.sin(5*x/2)

print(f'Root by bisection method = {bisection_root(f,[1.5,3],accur=10**-6)}')
print(f'Root by regula falsi method = {regula_falsi_root(f,[1.5,3],accur=10**-6)}')


x=np.linspace(1.5,3.5,1000)
y=[f(i) for i in x]
plt.plot(x,y,c='g',label='y=f(x)')
plt.axvline(x=2.555965304374695,c='r',label='root by bisection method')
plt.axvline(x=2.5559650927562716,c='orange',linestyle='--',label='root by regula falsi method')
plt.axhline(y=0)
plt.legend()
plt.grid(True)
plt.show()
```

# Question 2


```python
def find_bracket(f,bracket,beta):
    a=bracket[0]
    b=bracket[1]
    if f(a)*f(b)<=0:
        return [a,b]
    else:
        if abs(f(a))<=abs(f(b)):
            return find_bracket(f,[a-beta*(b-a),b],beta)
        else:
            return find_bracket(f,[a,b+beta*(b-a)],beta)
        
def f2(x):
    return -x-np.cos(x)

print(f'A valid bracket is: {find_bracket(f2,[2,4],0.5)}')

print(f'The root by bisection method is: {bisection_root(f2,[-2,4],10**-6)}')

    
```
