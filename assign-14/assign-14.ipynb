### Question-1

```python
def f2(y_vec,x):
    y=y_vec[0]
    return [(x+y)**2]
t_analytical=np.linspace(0,np.pi/5,100)
sol_analytical=[np.tan(x + np.pi/4) - x for x in t_analytical]
for i in [0.1,0.25,0.45]:
    sol_1 = RK4_solve(f2,[1],[0,np.pi/5],step_size=i)
    x_1 = sol_1[0]
    y_1 = [sol_1[1][i][0] for i in range(len(sol_1[1]))]
    plt.plot(x_1,y_1,label=f'step_size = {i}',linewidth=1)
plt.plot(t_analytical,sol_analytical,label='analytical expression',linewidth=2)
plt.xlabel('x')
plt.ylabel('y(x)')
plt.ylim(-0.5,20)
plt.legend()
plt.title('Solution for dy/dx = (x+y)^2')
plt.grid(True)
plt.show()
```

### Question-2

```python
#parameters
k=1
m=1
mu=0.15
omega_sqr=m/k


def f2(x_vec,t):
    x=x_vec[0]
    y=x_vec[1]

    return [y, -mu*y-omega_sqr*x]

sol=RK4_solve(f2,[1,0],[0,40])
t=sol[0]
x=[sol[1][i][0] for i in range(len(sol[1]))]
v=[sol[1][i][1] for i in range(len(sol[1]))]
E=[0.5*k*x[i]**2 + 0.5*m*v[i]**2 for i in range(len(x))]
plt.plot(t,x,label='x(t)',linewidth=1)
plt.plot(t,v,label='v(t)',linewidth=1)
plt.plot(t,E,label='E(t)',linewidth=1)
plt.xlabel('t')
plt.legend()
plt.grid(True)
plt.show()

plt.plot(x,v,linewidth=1)
plt.xlabel('x')
plt.ylabel('v')
plt.grid(True)
plt.title('Phase portrait')
plt.show()


```

### Lorentz System

```python
sigma=10
rho=28
beta=8/3

def dxdt(x_vec,t):
    x=x_vec[0]
    y=x_vec[1]
    z=x_vec[2]
    return [sigma*(y-x), x*(rho-z), x*y - beta*z]

lorentz_sol = RK4_solve(dxdt,[1,1,1],[0,40],step_size=0.01)

x_lor=[lorentz_sol[1][i][0] for i in range(len(lorentz_sol[1]))]
y_lor=[lorentz_sol[1][i][1] for i in range(len(lorentz_sol[1]))]
z_lor=[lorentz_sol[1][i][2] for i in range(len(lorentz_sol[1]))]

# fig = plt.figure(figsize=(8,6))
# ax = fig.add_subplot(projection='3d')
# ax.plot(x_lor, y_lor, z_lor, color='tab:blue', linewidth=0.5)
# ax.set_xlabel('x')
# ax.set_ylabel('y')
# ax.set_zlabel('z')
# ax.set_title('Lorenz attractor')
# ax.grid(True)
# plt.show()



```

```python
n=4
x_i = z_lor[:-n]
x_ii = x_lor[int(n/2):-int(n/2)]
x_iii = z_lor[n:]
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(projection='3d')
ax.plot(x_lor, y_lor, z_lor, color='tab:blue', linewidth=0.5)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Lorenz attractor')
ax.grid(True)
ax.plot(x_ii,x_i,x_iii,color='tab:red', linewidth=0.5)
# plt.plot(x_lor,y_lor)
# for i in range(len(x_lor) - 3):
#     ax.scatter(x_lor[i], x_lor[i+1], x_lor[i+2], c='r', s=5)

plt.show()
```
