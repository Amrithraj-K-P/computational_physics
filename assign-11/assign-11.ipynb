### Question-1

```python
def f1(x):
    return 1/x

def f2(x):
    return x*np.cos(x)

f1_m = midpoint_integ_accur(f1,[1,2],10**-7,1)
f1_s=simpson_integ_accur(f1,[1,2], 10**-7,1)

print('(value of the integral,n) for f1(x) = 1/x is:')
print(f'by midpoint method :{f1_m}')
print(f'By Simpsons method :{f1_s}')
print('')

f2_m=midpoint_integ_accur(f2,[0,np.pi/2], 10**-7, 2.29903)
f2_s=simpson_integ_accur(f2,[0,np.pi/2], 10**-7, 4.20257)
print('(value of the integral,n) for f2(x) = xcos(x) is:')
print(f'by midpoint method :{f2_m}')
print(f'By Simpsons method :{f2_s}')
```

## Question-2

```python
def f3(x):
    return np.sin(x)**2



n=np.arange(1,10000,100)
int_val=[monte_carlo_integ(f3,[-1,1],i)[0] for i in n]
plt.plot(n,int_val)
plt.axhline(y=0.545351286587,label='Actual value of integral', c='g')
plt.xlabel('N')
plt.ylabel('value of integral')
plt.ylim(0.5,0.6)
plt.grid(True)
plt.legend()
plt.show()

print(f'The value of the integral accurate upto 3 decimal places is : {np.average(int_val[int((3/4)*len(int_val)//1)::])}')



```
