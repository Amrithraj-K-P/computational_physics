```python
from Amrith_lib import *
import matplotlib.pyplot as plt
```

###   $dy/dx = y - x^2$

```python

def f1(y_vec,x):
    y=y_vec[0]
    return [y-x**2]

def f2(y_vec,x):
    y=y_vec[0]
    return [(x+y)**2]
sol=forward_euler_solve(f1,[0],[0,2])
sol_pred_corr=predictor_corrector_solve(f1,[0],[0,2])
sol_analytical=[x**2 + 2*x +2 -2*np.exp(x) for x in sol_pred_corr[0]]




plt.plot(sol[0],[sol[1][i][0] for i in range(len(sol[1]))],label='forward Euler')
plt.plot(sol_pred_corr[0],[sol_pred_corr[1][i][0] for i in range(len(sol_pred_corr[1]))],label='Predictor_corrector')
plt.plot(sol_pred_corr[0], sol_analytical,label='analytical result')
plt.xlabel('x')
plt.ylabel('y(x)')
plt.legend()
plt.title('Solution for dy/dx = y - x^2')
plt.grid(True)
plt.show()
```

### $dy/dx = (x+y)^2$

```python
sol2=forward_euler_solve(f2,[1],[0,np.pi/5],step_size=0.01)
sol2_pred_corr=predictor_corrector_solve(f2,[1],[0,np.pi/5],step_size=0.01)
sol_2_analytical=[np.tan(x + np.pi/4) - x for x in sol2_pred_corr[0]]


plt.plot(sol2[0],[sol2[1][i][0] for i in range(len(sol2[1]))],label='forward Euler')
plt.plot(sol2_pred_corr[0],[sol2_pred_corr[1][i][0] for i in range(len(sol2_pred_corr[1]))],label='Predictor_corrector')
plt.plot(sol2_pred_corr[0], sol_2_analytical,label='analytical result')
plt.xlabel('x')
plt.ylabel('y(x)')
plt.legend()
plt.title('Solution for dy/dx = (x+y)^2')
plt.grid(True)
plt.show()
```
