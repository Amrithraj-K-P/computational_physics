# Question - 2

```python
x2=[2.5,3.5,5,6,7.5,10,12.5,15,17.5,20.5]
y2=[13,11,8.5,8.2,7,6.2,5.2,4.8,4.6,4.3]

log_y2 = [float(np.log(i)) for i in y2]
log_x2 = [float(np.log(i)) for i in x2]

rsqr_power_law = lsqfit(log_x2,log_y2)[2]
r_sqr_exponential = lsqfit(x2,log_y2)[2]

print(f'r^2 for the power law is: {rsqr_power_law}')
print(f'r^2 for the exponential law is: {r_sqr_exponential}')

print('Hence the power law gives better fit')
```
